```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("data_cleaned.csv")
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['price_num'], kde=True, bins=30)
plt.title('Car Price Distribution')
plt.xlabel('Price (VND)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['mileage_num'], kde=True, bins=30, color='orange')
plt.title('Mileage Distribution')
plt.xlabel('Mileage (km)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
brand_mean_price = df.groupby("Brand")["price_num"].mean().sort_values(ascending=False)
sns.barplot(x=brand_mean_price.index, y=brand_mean_price.values)
plt.title('Average Price by Brand')
plt.ylabel('Average Price (VND)')
plt.xlabel('Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Fuel_Type', palette='Set2')
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Year', y='price_num')
plt.title('Price Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Price (VND)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='mileage_num', y='price_num', hue='Brand', alpha=0.7)
plt.title('Price vs Mileage')
plt.xlabel('Mileage (km)')
plt.ylabel('Price (VND)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Transmission', palette='pastel')
plt.title('Number of Cars by Transmission Type')
plt.xlabel('Transmission')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
top_provinces = df['Province'].value_counts().nlargest(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_provinces.index, y=top_provinces.values)
plt.title('Top 10 Provinces by Number of Car Listings')
plt.xlabel('Province')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Body_Type', y='price_num', estimator='mean')
plt.title('Average Price by Body Type')
plt.xlabel('Body Type')
plt.ylabel('Average Price (VND)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
numeric_cols = ['Year', 'price_num', 'mileage_num']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Fuel_Type', y='price_num')
plt.title('Price Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Price (VND)')
plt.tight_layout()
plt.show()
```
